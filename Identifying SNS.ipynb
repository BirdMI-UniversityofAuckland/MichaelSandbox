Here we want to identify segments of song and non-song.
We will aim to do this by first identifying segments of 'song' by taking any instance of sound longer than 500 ms, with no periods of silence greater than 20 ms.
Then we will identify periods of silence, by finding silent intervals with no sound for a length longer than 100 ms.

These criteria are designed to best capture and exclude our desired datasets.

We will start by importing our dependancies, and also defining some constants



```python
#Here we are simply importing necessary methods from the library of definitions created.
%run 'Mibrary.ipynb' 
#We also wish to see our graphs within the notebook - not as a separate window instance.
%matplotlib inline 

Fs = 44100.0 # Sample rate, 44100 Hz

min_length_of_song = 0.500 * Fs #500 ms, any sounds less than this length will not qualify as song.
min_length_of_silence = 0.200 * Fs # 100ms, any silences less than this length will not qualify as non-song.
max_inter_syllable_gap = 0.020 * Fs # 20ms, the maximum length of silence allowed in a song, represents the inter-syllable gaps.

```

```python
greater_dir = "C://Users//mbar372//Documents//FILTER_TEST_SAMPLES//raw_data//"

"""
The data are now stored as such:

data_array [ X ] [ Key ]
where X is the sample number ( 0 - 7 ) i.e. 8 samples
and Key stands for either 'sng' , 'e1', or 'e2'.
Implicating a call for either the song, electrode 1, or electrode 2's
data array over the entire length of the afore-choosen sample.

Thus to access the 5th samples 2nd electrode we simple use:
wanted_data = data_array[4]['e2']
"""

files = os.listdir(greater_dir )
data_array = []
for f in files:
    datum = load_mat_file(greater_dir + f)
    cleaned_datum = {'id': f,
                     'sng': np.array(datum['sng'][:,0], dtype = 'float64') ,
                     'e1': np.array(datum['e1'][0,:], dtype = 'float64'),
                     'e2': np.array(datum['e2'][0,:], dtype = 'float64'),
                     'ref': np.array(datum['ref'][0,:], dtype = 'float64')}
    data_array.append(cleaned_datum)
```

The first thing we must do, is filter and rectify the data, to convert it to an amplitude envelope, as this is the envelope we will be using to judge a threshold of what is loud enough to be sound, or not sound.

We will simply perform our initial tests on the first example of sample song , which we will view below

```python
# We will work with the first sample to start with
our_data = data_array[1]['sng']
t = np.linspace(0,len(our_data)/Fs,len(our_data))
plt.plot(t,our_data)
plt.title(data_array[0]['id'])
plt.xlabel("time (s)")
plt.ylabel("Amplitude (V)")
plt.show()
```

We will rectify the data, and then use the Root Mean Square linear Filter to smooth it, thus forming our amplitude envelope. We will also take this opportunity to convert our signal into Decibels.

```python
# Calculate the time axis for visualisation purposes only.
t = np.linspace(0,len(our_data)/Fs,len(our_data))

# Rectify
data_rect = abs(our_data)
# We take the rootmean squared of the signal ,
# using a largish window size of 250 ms works better for the desired smoothing in this application.
data_rect_filtrate = linear_filt_rms(data_rect, 0.25 * Fs )
# We also convert the data into dB
data_rect_filtrate = 10 * np.log10(data_rect_filtrate)

plt.plot(t,data_rect_filtrate)
plt.title("Amplitude Envelope")
plt.xlabel("time (s)")
plt.ylabel("Amplitude (V)")
plt.show()

plt.specgram(our_data, Fs = 44100)
plt.title("Spectrogram")
plt.xlabel("time (s)")
plt.ylabel("Amplitude (V)")
plt.show()
```

Now that we have an amplitude envelope to work with, we can try to apply some of the criteria we have established to it, we will loop through each position, holding a count of how long it has been since silence was not heard. If we last longer than 500 ms without a silence greater than 20 ms, we will count this as our first example of data.

```python
new_data = data_rect_filtrate

threshold = 27 # we set the threshold to 27 dB

output = find_songs(new_data, min_length_of_song, max_inter_syllable_gap, threshold, draw = 'yes')
```

Great! We now have a working 'sound finder' where we only need to change the variables to get what outcome we want , in particular, threshold is the largest contributor here , and choosing an appropriate thresholding value is crucial to having any kind of reliable output . Although our dB scale did not callibrate, it does not need do , as all these recordings are assumed to have been recorded by a callibrated microphone. What we do see throughout all recordings is that 'silence' hangs around 25 or 26 dB or less, whereas the height of songs can go > 30 dB on this scale. What is important is the relative difference between these two points. 

Due to this lack of callibration , and likely variances and differences between all the sound files anyway, it becomes ideal to have some form of defined thresholding reasoning, that can be automated,  rather than a fixed value, or having to go through each one by eye, and choosing the best amplitude threshold manually.

Initial ideas are these:

Take the mean of the signal, and set the threshold at that minus a standard deviation %, this could be applied to all datasets equally.
The downside to the above , is that any dataset that has little to no sound / song, will experience poorer selection (less variance within the data set thus a smaller std dev to be removed from the mean) ... It is therefore preferable to also have a minimum value that the resulting threshold calc can not go below.

Preliminary testing has shown that the following logical expression:
Threshold =  Maximum of eiher (Mean - Stdev ) or (25 dB) 
Has worked well in the test samples , responding to variation of both heavily song concentrated datasets , and those with little, no, or noisy sounds instead of song where the variance is less. 

Another idea is to normalise each dB , although this would run into similar problems for songs that have less ampltiude.

Finally, it may be best to pick a dB across all song datasets, that is the mean - std of all datasets. That could give us a more 'universal' way of determining an effecting threshold.




```python
# It can help to visualise our segregations on the original spectrogram of the sound.

plt.specgram(our_data,Fs = 44100, scale_by_freq = "false",  scale = "dB")
for n in range(len(output)):
    point1 = output[n][1]
    point2 = output[n][2]
    plt.axvline(point1/Fs, linewidth=2, color='g')
    plt.axvline(point2/Fs, linewidth=2, color='r')
plt.show()

print "***** ***** ***** ***** ***** ***** ***** ***** *****"
```
