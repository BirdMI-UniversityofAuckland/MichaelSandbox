```python
 ## Creates an array containing numpy arrays of all wav files from a directory, processed to be normalised

def create_song_dataset(directory): # input is the working directory where 'rawsong' folder is kept
    
    ## This is an array of directories to the song files to be tested.
    raw_song_dir_array  = glob.glob(directory + '\\*.wav')

    ## This cycles through every wav file in the array and assigns its 'data' ( [1] is used to return data only) to song_array
    song_array = [wav.read(n)[1] for n in raw_song_dir_array]
    
    ## Normalises each song individually before creating final output array
    song_array = [pcm2float(n,'float64') for n in song_array]
    
    return song_array ## Return array of numpy arrays filled with normalised song data.
```

```python

# NOTE - Took this from GitHub user mgeier - see GitHub repository where fork has been made.

def pcm2float(sig, dtype='float64'):
    """Convert PCM signal to floating point with a range from -1 to 1.
    Use dtype='float32' for single precision.
    Parameters
    ----------
    sig : array_like
        Input array, must have (signed) integral type.
    dtype : data type, optional
        Desired (floating point) data type.
    Returns
    -------
    ndarray
        normalized floating point data.
    See Also
    --------
    float2pcm, dtype
    """
    # TODO: allow unsigned (e.g. 8-bit) data

    sig = np.asarray(sig)
    if sig.dtype.kind != 'i':
        raise TypeError("'sig' must be an array of signed integers")
    dtype = np.dtype(dtype)
    if dtype.kind != 'f':
        raise TypeError("'dtype' must be floating point type")

    # Note that 'min' has a greater (by 1) absolute value than 'max'!
    # Therefore, we use '-min' here to avoid clipping.
    return sig.astype(dtype) / dtype.type(-np.iinfo(sig.dtype).min)
```

```python
def correlate_coef(array1,array2):
    """This Definition takes two arrays of sound data, and calculates an allignment for maximum correlation,
    it then adds this offset , plus additional padding to ensure that the two arrays are the same size,
    this allows for the two arrays to be compared using the numpy coefficient function. This is then used to
    return a percentage value of the overal similarity between the two sequences. The Percentage value is only
    taking into account the overlapping regions of the two padded arrays.
    """
    #array2 = np.pad(array2, int( array1.shape[0]) , mode = 'constant' , constant_values = 0 )
    
    #sizes and calculate the padding value
    size1 = int(array1.shape[0])
    size2 = int(array2.shape[0])
    #store original sizes of real data for future use after padding occurs.
    orig_size1 = size1
    orig_size2 = size2
    
    print orig_size1
    print orig_size2
    
    # Calculate padding necesarry to make size1.len = size2.len
    padding_value = abs( size1 - size2 )
    
    ## Draw out what is about to be padded:
    #print "Amplitude data for raw Song:"
    #draw_wav_output(array1, spectrogram = True)
    #print "Amplitude data for bmi Song:"
    #draw_wav_output(array2, spectrogram = True)

    ######################## This has been made redundant, as it wasn't working well and FFT is way cooler / better
    # Cross Correlate the two arrays, return an offset value, indicating offset of maximum similarity
    #corrdata = np.correlate(array1,array2)
    #offset = int(np.where(corrdata == corrdata.max())[0][0]) # Ensures offset is an int, not int64 of numpy
    #print "Max Correlation shows using an offset of " + str(offset) +":"
    #draw_wav_output(corrdata, lineweight=0.5)
    
    #pad the smallest array's end to be the same size as the larger.
    print "Padding by difference between sizes: " + str(padding_value)
    if size1<size2:
        array1 = np.pad(array1, (0,padding_value), mode = 'constant', constant_values = 0)
    if size2<size1:
        array2 = np.pad(array2, (0,padding_value), mode = 'constant', constant_values = 0)

    # To check lengths are the same still, and update to new values.
    size1 = int(array1.shape[0])
    size2 = int(array2.shape[0])
    print "Difference between shapes is, " + str(size1) + " and " + str(size2) + "." 
    
    ## Calculate a new offset using FFT
    A = fftpack.fft(array1)
    B = fftpack.fft(array2)
    Ar = -A.conjugate()
    Br = -B.conjugate()
    offset1test = np.argmax(np.abs(fftpack.ifft(Ar*B)))
    offset2test = np.argmax(np.abs(fftpack.ifft(A*Br))) ## This is simple the overshoot of offset1 past the length of the array2
    print "Test1: " + str(offset1test) + ", Test2: " + str(offset2test) + "Test2-size: " + str(offset2test-orig_size2)
    
    if offset1test < orig_size1:
        offset = offset1test
    else:
        offset = offset2test
        
    
    ## Pad by the offset calculated by the correlation onto the beginning of the first array, and onto the end of the second.
    print "Padding by offset: " + str(offset)
    array1 = np.pad(array1, (int(offset),0), mode = 'constant', constant_values = 0)
    array2 = np.pad(array2, (0,int(offset)), mode = 'constant', constant_values = 0)
    
    # Calculate the points of overlap between the two arrays
    point1 = offset
    point2 = min(offset+orig_size1,orig_size2) 
    
    ## Draw out what is about to be correlated:
    print "Amplitude data for raw Song:"
    draw_wav_output(array1, spectrogram = True)
    print "Amplitude data for bmi Song:"
    draw_wav_output(array2, spectrogram = True)
    
    plt.plot(array1)
    plt.plot(array2-0.01)
    plt.axvline(point1+1, linewidth=2, color='r')
    plt.axvline(point2-1, linewidth=2, color='r')
    plt.show()
    
    comparearray1 = array1[point1:point2]
    comparearray2 = array2[point1:point2]

    #corrdata = remove_negative(corrdata)
    print "Final Coefficients Data:"
    corrcoefdata = np.corrcoef(comparearray1,comparearray2)
    #draw_wav_output(corrcoefdata,lineweight=1, spectrogram = False)

    ## Print out the Coefficient value 
    coefficient_value = abs(int(corrcoefdata[1,0]*100000)/1000.00)
    print coefficient_value
    

    
    return coefficient_value
```
