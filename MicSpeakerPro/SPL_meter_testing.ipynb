```python
def quick_view_sound(sound_data, pcm = False, Fs = 44100):
    """ Helper function to aid in testing, give it the sound (if pcm, set pcm to True) , and it will quickly 
    draw a spectrogram and raw trace within the same plot size to get a quick idea of what the sound is without having
    to play it.
    """
    if pcm == True:
        sound_data = pcm2float(sound_data, dtype = 'float32')
        
    plt.figure(1)
    plt.subplot(211)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.plot(sound_data, c= 'r')

    plt.subplot(212)
    plt.xlabel('Time')
    plt.ylabel('Frequency')
    plt.specgram(sound_data, NFFT = 512, Fs = Fs)
    plt.show()
    
    
def create_comparison_arrays(array1,array2, draw = "no", force_align = 0):
    """The purpose of this function is to take two arrays, padd them to the same size, 
    find the point of highest similarity using conjugates in the Fourier plane, 
    and then offset these padded arrays by an amount to create two comparable datasets at a point of highest similarity.
    
    Additionally the user can pass force_align with a sample value, to manually assign an alignment.
    An alignment of 0 will re-calculate a new best alignment and apply it.
    """    
    #sizes and calculate the padding value
    size1 = len(array1)
    size2 = len(array2)
    
    #store original sizes of real data for future use after padding occurs.
    orig_size1 = size1
    orig_size2 = size2
        
    # Calculate padding necesarry to make size1.len = size2.len
    padding_value = abs( size1 - size2 )
   
    #pad the smallest array's end to be the same size as the larger.
    #print "Padding by difference between sizes: " + str(padding_value)
    if size1<size2:
        array1 = np.pad(array1, (0,padding_value), mode = 'constant', constant_values = 0)
    if size2<size1:
        array2 = np.pad(array2, (0,padding_value), mode = 'constant', constant_values = 0)

    # To check lengths are the same still, and update to new values.
    size1 = len(array1)
    size2 = len(array2)
    if size1 != size2:
        #print "Difference between shapes is, " + str(size1) + " and " + str(size2) + "." 
        print "Error, the two arrays failed to pad properly."
    
    ## Here we check, have we been provided with alignment? If no, then calculate one...
    if force_align == 0 : 
        ## Calculate a new offset using FFT
        A = fftpack.fft(array1)
        B = fftpack.fft(array2)
        Ar = -A.conjugate()
        Br = -B.conjugate()
        offset1test = np.argmax(np.abs(fftpack.ifft(Ar*B)))
        offset2test = np.argmax(np.abs(fftpack.ifft(A*Br))) ## This is simply the overshoot of offset1 past the length of the array2
    
    ## Depending on which offset is shorter, depends on which way to move the arrays relative to each other to achieve maximum correlation.
        if offset1test < offset2test:
            offset1 = offset1test
            offset2 = 0
        else:
            offset1 = 0
            offset2 = offset2test
        
    ## Pad by the offset calculated by the correlation onto the beginning of the first array, and onto the end of the second.
    ## Note this is calculated depending on the smallest movement, and is thus applied as either offset 1 or offset 2, but switching
    ## which dataset to add it to the front or back of.
        array1 = np.pad(array1, (int(offset1),int(offset2)), mode = 'constant', constant_values = 0)
        array2 = np.pad(array2, (int(offset2),int(offset1)), mode = 'constant', constant_values = 0)
        
    ## If we have been given an offset to use instead of calculating, use that offset instead of going through calculating our own.
    else:
        offset1 = 0
        offset2 = 0
        if force_align > 0:
            offset1 = force_align
        else:
            force_align < 0
            offset2 = abs(force_align)
        
        array1 = np.pad(array1, (int(offset1),int(offset2)), mode = 'constant', constant_values = 0)
        array2 = np.pad(array2, (int(offset2),int(offset1)), mode = 'constant', constant_values = 0)
        
    # Calculate the points of overlap between the two original arrays, these will be used to clip excess padding.
    point1 = offset1 + offset2
    point2 = min(offset1 + orig_size1, offset2 + orig_size2)
    
    # If draw is yes, then draw the plots to visualise what has occured...
    if draw == 'yes':
        plt.figure(1)
        plt.xlabel('Time')
        plt.ylabel('Frequency')
        plt.subplot(211)
        plt.plot(array1)
        plt.axvline((point1+1), linewidth=2, color='r')
        plt.axvline((point2-1), linewidth=2, color='r')
        plt.subplot(212)
        plt.plot(array2)
        plt.axvline((point1+1), linewidth=2, color='r')
        plt.axvline((point2-1), linewidth=2, color='r')
        plt.show()
    
    comparearray1 = array1[point1:point2]
    comparearray2 = array2[point1:point2]

    # Return only the maximum overlap of the two arrays at the maximum correlation point offset.
    return comparearray1, comparearray2, offset1-offset2


"""
=====================================================================================================
We've got this from audio processing github repo, handy for making conversions between datatypes.
https://github.com/mgeier/python-audio/blob/master/audio-files/utility.py

"""

import numpy as np
import contextlib

def float2pcm(sig, dtype='int16'):
    """Convert floating point signal with a range from -1 to 1 to PCM.
    Parameters
    ----------
    sig : array_like
        Input array, must have floating point type.
    dtype : data type, optional
        Desired (integer) data type.
    Returns
    -------
    ndarray
        integer data.
    See Also
    --------
    pcm2float, dtype
    """
    # TODO: allow unsigned (e.g. 8-bit) data

    sig = np.asarray(sig)
    if sig.dtype.kind != 'f':
        raise TypeError("'sig' must be a float array")
    dtype = np.dtype(dtype)
    if dtype.kind != 'i':
        raise TypeError("'dtype' must be signed integer type")

    return (sig * np.iinfo(dtype).max).astype(dtype)

def pcm2float(sig, dtype='float64'):
    """Convert PCM signal to floating point with a range from -1 to 1.
    Use dtype='float32' for single precision.
    Parameters
    ----------
    sig : array_like
        Input array, must have (signed) integral type.
    dtype : data type, optional
        Desired (floating point) data type.
    Returns
    -------
    ndarray
        normalized floating point data.
    See Also
    --------
    float2pcm, dtype
    """
    # TODO: allow unsigned (e.g. 8-bit) data

    sig = np.asarray(sig)
    if sig.dtype.kind != 'i':
        raise TypeError("'sig' must be an array of signed integers")
    dtype = np.dtype(dtype)
    if dtype.kind != 'f':
        raise TypeError("'dtype' must be floating point type")

    # Note that 'min' has a greater (by 1) absolute value than 'max'!
    # Therefore, we use '-min' here to avoid clipping.
    return sig.astype(dtype) / dtype.type(-np.iinfo(sig.dtype).min)

"""
end of github mgeier repo.
=====================================================================================================
"""


def pass_through_hardware(sound, fs = 44100, block_length = 1024):
    """
    Takes a sound, and plays and records it from the default recording and default audio devices. 
    PARAMETERS:
    -------------
    sound : array-like, linear float
        The sound that is to be played, and then recorded.

    fs : int
        The sampling frequency to use
        
    block_length : int
        The size of the blocks that samples sent to the sound-card, to play and be recorded in.
        
    RETURNS
    -------------
    output_wav : array-like, linear float
        The sound that was recorded from the sound-card.
    
    """
    
    output_wav = np.array(1) #Create an empty number array, this will be appended to with the microphone inputs

    # Create the stream instance (An object representing the sound card), and start it.
    s = Stream(sample_rate=fs, block_length=block_length)
    s.start()
    
    #Iterate over the length of the sound to be played with the block size,
    for i in range(len(sound)/block_length):
        # Note, there exists an issue, where it takes approximately ~6000 samples to start playback,
        # This is evident where 
                
        # PLAYBACK the next block of sound
        block_to_write = sound[i*block_length:i*block_length+block_length]
        s.write(block_to_write)
        
        # RECORD the next block of sound. Append to new Numpy array.
        new_block = s.read(block_length)[:,0]
        output_wav = np.append(output_wav, new_block)
  
    # Record the next 0.20s of sound (This is handy for addressing the possible undershoot recording problem discussed above)
    new_block = s.read(8810)[:,0]
    output_wav = np.append(output_wav, new_block)
    
    #Stop the stream
    s.stop()
    
    # Return the microphones experiences during sound playback.
    return output_wav


def update_test_lib(working_directory):
    fs = 44100
# CREATE ASCENDING TONE
    time_series = np.linspace(0, 10, 44100*10)
    ascendingtone = signal.chirp(time_series, f0=10.0, f1=20000.0, t1=10, method = 'log')
    ascendingtone = float2pcm(ascendingtone, dtype ='int16')
    wav.write(working_directory + 'TEST_LIB\\ascendingtone.wav',fs, ascendingtone)
# CREATE ASCENDING TONE LOOONNGGGGG
    time_series = np.linspace(0, 30, 44100*30)
    ascendingtone = signal.chirp(time_series, f0=20.0, f1=20000.0, t1=30, method = 'log')
    ascendingtone = float2pcm(ascendingtone, dtype ='int16')
    wav.write(working_directory + 'TEST_LIB\\ascendingtone_long.wav',fs, ascendingtone)
# CREATE PURE TONES.
    t = np.r_[0:int(fs*5):1] / float(fs)
    fnarray = [100,150,200,300,400,500,600,700,800,900,
               1000, 1500,2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000] #Array of frequencies to use.
    for n in range(len(fnarray)):
        fn = fnarray[n]
        puretone = np.sin(2 * np.pi * t * fn)
        puretone = float2pcm(puretone, dtype ='int16')
        wav.write(working_directory + 'TEST_LIB\\sound_'+ str(fnarray[n]) + '.wav',fs, puretone)
        
# Create silence
    silence = np.zeros(44100 * 10)
    silence = float2pcm(silence, dtype ='int16')
    wav.write(working_directory + 'TEST_LIB\\sound_silence.wav',fs, silence)
    
# Create single file pure tone with 1 second intervals
    finalsound = []
    t = np.r_[0:int(fs):1] / float(fs)
    for n in range(len(fnarray)):
        fn = fnarray[n]
        puretone = np.sin(2 * np.pi * t * fn)
        puretone=np.pad(puretone,22050,mode = 'constant', constant_values = 0)
        finalsound = np.append(finalsound,puretone)
    
    finalsound=np.pad(finalsound,22050,mode = 'constant', constant_values = 0)
    finalsound = float2pcm(finalsound, dtype ='int16')
    wav.write(working_directory + 'TEST_LIB\\sound_alltones.wav',fs, finalsound)

# Create single file 1000 Hz tone with 1 second intervals of increasing amplitude.
    finalsound = []
    t = np.r_[0:int(fs):1] / float(fs)
    amplitude_array = np.logspace(-1,1,20)/10
    for n in range(len(amplitude_array)):
        fn = fnarray[n]
        puretone = np.sin(2 * np.pi * t * 1000)*amplitude_array[n]
        puretone=np.pad(puretone,22050,mode = 'constant', constant_values = 0)
        finalsound = np.append(finalsound,puretone)
    
    finalsound=np.pad(finalsound,22050,mode = 'constant', constant_values = 0)
    finalsound = float2pcm(finalsound, dtype ='int16')
    wav.write(working_directory + 'TEST_LIB\\sound_amplitudes.wav',fs, finalsound)

def rms_voltage(a, window_size):
    """ Apply a Root Mean Squared averaging window across an array
    Inputs:
        a , array-like, linear dataset
        window_size , The size of window to convolve across the dataset
    Returns:
        The rolling windowed root mean sqaure of input a
    """
    a2 = np.power(a,2)
    window = np.ones(int(window_size))/float(window_size)
    return np.sqrt(np.convolve(a2, window, 'valid'))

def volts_to_db(array, dBoffset=94, window_size = 5512):
    """
    A Simple helper function, this takes the voltage output raw trace capture from the microphone input (SPL meter)
    and filters it with a rolling window , currently taking max value, and then converts it into dB
    NOTE the dB conversion depends on using the correct offset.
    In 50-100 dB Mode OR 30-80 dB mode on the SPL meter, a 94 dB offset is appropriate, 
    Oddly enough , ~94 dB = 1 Pascal [93.9794 dB SPL], which seems very coincidental 
    especially considering SPL is effectively measuring Pascals.
  
    In a 30 - 130 dB Mode (Auto mode), an offset of 74 is appropriate.
    I have not generated a sound large enough to use the 80-120 dB feature, although I would predict an offset of 114 dB
    The offset can be checked by comparing the output of this to the raw readings on the display of the SPL meter.
    
    Furthermore, and this looks to be a final proof that such an offset is appropraite , is the the calibration numbers given as settings:
    http://www.jaycar.co.nz/Test-%26-Measurement/Environmental-Meters/Sound-Meters/Pro-Sound-Level-Meter-with-Calibrator/p/QM1592
    "Calibration: 94, 114dB"
    
    
    Inputs:
        array - raw sound wave array
    
        dBoffset - The offset in dB to apply to the Voltage signal
    
    Outputs:
        array-like - the raw sound converted to dB SPL if the correct offset was used.
        
    """
    array_out = 20*np.log10(rms_voltage(array,window_size)) + dBoffset

    
    return array_out


def rolling_window(a, window):
    """
    Creates an array of windows for each position that can then be used to be worked on.
    Uses C libraries instead of for loops for fasterness.
    
    OUTDATED, not longer used here as the above methods are faster.
    """      
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)


"""
Let's create a C Weighting dB curve!!! , we can use this to ensure that the weighting we recieve is dB SPL
The equation for C Weighting is:
C(f) = (12200^2 * f^2) / ( (f^2 * 20.6^2) * (f^2 * 12200^2))
"""

def c_to_spl(f, dB = 0):
    """ given a frequency, this function returns the offset
        for the C weighted curve. If you provide a dB argument, it will
        perform the conversion        
    """

    c_offset = float((12200.0**2) * (f**2)) / float( ((f**2) + (20.6**2)) * ((f**2) + (12200.0**2)))
    c_offset_db = (20.0 * np.log10(c_offset)) + 0.06 # We apply a small constant offset here too as this is part of the conversion to C
    
    return  dB - c_offset_db # We minus to remove the offset imposed by the C weighting at the given frequency.
```
