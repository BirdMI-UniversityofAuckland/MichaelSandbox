```python
def rolling_window(a, window):
    """
    Creates an array of windows for each position that can then be used to be worked on.
    Uses C libraries instead of for loops for fasterness.
    It also pads edges with 0's so that the returned function is the same shape as the input.
    Also it alters the size of the window if it is odd to achieve this.
    """
    if window%2 != 0:
        window+=2
    else:
        window+=1
    
    a = np.pad(a,(window/2,window/2), mode = 'mean')
    
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    
    strides = a.strides + (a.strides[-1],)
    
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
```

```python
def linear_filt_rms(array, window_size):
    """ Apply a Root Mean Squared averaging window across an array linearly
    Inputs:
        a , array-like, linear dataset
        window_size , The size of window to convolve across the dataset
    Returns:
        The rolling windowed root mean sqaure of input a
    """
    a2 = np.power(array,2)
    window = np.ones(int(window_size))/float(window_size)
    return np.sqrt(np.convolve(a2, window, 'same'))

def linear_filt(array, window_size):
    """ Apply averaging window across an array linearly
    Inputs:
        a , array-like, linear dataset
        window_size , The size of window to convolve across the dataset
    Returns:
        The rolling windowed root mean sqaure of input a
    """
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(array, window, 'same')
```

```python
"""
=====================================================================================================
We've got this from audio processing github repo, handy for making conversions between datatypes for AUDIO.
https://github.com/mgeier/python-audio/blob/master/audio-files/utility.py

"""

import numpy as np
import contextlib

def float2pcm(sig, dtype='int16'):
    """Convert floating point signal with a range from -1 to 1 to PCM.
    Parameters
    ----------
    sig : array_like
        Input array, must have floating point type.
    dtype : data type, optional
        Desired (integer) data type.
    Returns
    -------
    ndarray
        integer data.
    See Also
    --------
    pcm2float, dtype
    """
    # TODO: allow unsigned (e.g. 8-bit) data

    sig = np.asarray(sig)
    if sig.dtype.kind != 'f':
        raise TypeError("'sig' must be a float array")
    dtype = np.dtype(dtype)
    if dtype.kind != 'i':
        raise TypeError("'dtype' must be signed integer type")

    return (sig * np.iinfo(dtype).max).astype(dtype)

def pcm2float(sig, dtype='float64'):
    """Convert PCM signal to floating point with a range from -1 to 1.
    Use dtype='float32' for single precision.
    Parameters
    ----------
    sig : array_like
        Input array, must have (signed) integral type.
    dtype : data type, optional
        Desired (floating point) data type.
    Returns
    -------
    ndarray
        normalized floating point data.
    See Also
    --------
    float2pcm, dtype
    """
    # TODO: allow unsigned (e.g. 8-bit) data

    sig = np.asarray(sig)
    if sig.dtype.kind != 'i':
        raise TypeError("'sig' must be an array of signed integers")
    dtype = np.dtype(dtype)
    if dtype.kind != 'f':
        raise TypeError("'dtype' must be floating point type")

    # Note that 'min' has a greater (by 1) absolute value than 'max'!
    # Therefore, we use '-min' here to avoid clipping.
    return sig.astype(dtype) / dtype.type(-np.iinfo(sig.dtype).min)

"""
end of github mgeier repo.
=====================================================================================================
"""
```

```python
def normalise(data):
    """
    Just a quick and handy definition that will normalise its input data by its maximum absolute value,
    The output is as a numpy array.
    """
    return np.array(data / np.max(abs(data)), dtype = 'float64')
```

```python
def find_songs(data, min_length, max_silence, threshold, draw = 'no'):
    """
    This method finds lengths of a minimum length, above a threshold,
    and that have no silences longer than a minimum silence value.
    
    Once it finds these snippets of sound, it returns then in an appendent list,
    the output also includes start and stop sample ID's for each snippet extracted.
    
    user inputs;
    
     - data,   array-like, the sound data.
     
     - min_length,   Integer, The minimum length in number of samples,
                     a sound has to be, before it is extracted.
                     
     - max_silence,  Integer, the maximum length of samples below threshold 
                     tolerable to exist in the extracted sound.
                     
     - Threshold, array-like, same length as data, represents what the threshold
                  value is at each sample point
                     
     - draw,   String, accepts ['yes'||'no']  Indicates whether to draw the outcome in matplotlib.
               The outcome is drawn as a the raw data, with horizontal lines inticating the start (green)
               and end (red) of each segment found from the raw data.
                     
    Outputs;
    
     - array[N , 3],   array-like with shape [N, 3] where N is 0 to infinity, for each song extracted, 
                       for the second axis, 0 is the data, 1 is the start sample, and 2 is the final sample

    """
    
    sound_count = 0
    silence_count = 0
    sound_start = 0
    capture = 0
    sound_data = []

    for i in range(len(data)):
        # If we are above threshold, count this sample as a sound.
        if data[i] > threshold[i]:
            sound_count +=1
            silence_count = 0
        
            # Document this as the start of a sound if we are not already in one.
            if sound_start == 0:
                sound_start = i
    
        # If we are below threshold, count this sample as a silence.
        else:
            silence_count+=1
    
        # If our silence count has reached more than our minimum requirement, reset our sound counting.
        if silence_count > max_silence:
        
            # but only reset after saving whether or not we have a sound >500 ms.
            if capture == 1:
                if draw == 'yes':
                    print "A sound was saved from " + str(sound_start) + " to " + str(i) + "."
                sound_end = i - max_silence # Here remember to remove our ending criteria of the song. 
                sound_data.append([data[sound_start:sound_end],sound_start,sound_end])
                capture = 0
        
            sound_count = 0
            sound_start = 0
    
        if sound_count >= min_length:
            capture = 1
    
    # Remembering that just because we have reached the end of the file, doesn't mean we should be discounting if we are in a song.
    # This does not happen often.
    if capture == 1:
        if draw == 'yes':
            print "A sound was found from " + str(sound_start) + " to " + str(i) + "."
        sound_end = i - max_silence # Here remember to remove our ending criteria of the song. 
        sound_data.append([data[sound_start:sound_end],sound_start,sound_end])
        capture = 0
        
    if draw == 'yes':
        plt.plot(t, data)
        plt.plot(t, threshold, linewidth=2, color='k')

        for n in range(len(sound_data)):
            point1 = sound_data[n][1]
            point2 = sound_data[n][2]

            plt.axvline(point1/Fs, linewidth=2, color='g')
            plt.axvline(point2/Fs, linewidth=2, color='r')
    
        plt.show()
        
    return sound_data
```
