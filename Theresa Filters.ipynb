```python
""" Some basic Constants """
# Sampling Frequency
Fs = 44100.0

# Filter time-window size of 50 ms
window_size = Fs * 0.05 

""" Savitzky-Golday """
# FILTER CONSTANTS
SG_order = 4

def apply_SG_FILTER(data, order, window_size = window_size):
    SG_filtrate = signal.savgol_filter(data, window_size, order, deriv=0, delta=1.0, axis=-1, mode='interp', cval=0.0)
    return SG_filtrate

""" Butter-worth """
# FILTER CONSTANTS
low_pass = 100
high_pass = 5000
bandtype = 'lowpass'
BW_order = 4

def apply_BW_FILTER(data, low_pass, high_pass, bandtype, order):
    start_fs = (low_pass) / (Fs / 2.0)
    stop_fs = (high_pass) / (Fs / 2.0)
    
    
    if bandtype == 'lowpass':
        critical_fs = start_fs
    if bandtype == 'highpass':
        critical_fs = stop_fs
    if bandtype == 'bandpass':
        critical_fs = [start_fs, stop_fs]

    b, a = signal.butter(4, critical_fs, btype= bandtype)
    BW_filtrate = signal.filtfilt(b, a, data)
    
    return BW_filtrate
```

```python
greater_dir = "C://Users//mbar372//Documents//FILTER_TEST_SAMPLES//raw_data//"

"""
The data are now stored as such:

data_array [ X ] [ Key ]
where X is the sample number ( 0 - 7 ) i.e. 8 samples
and Key stands for either 'sng' , 'e1', or 'e2'.
Implicating a call for either the song, electrode 1, or electrode 2's
data array over the entire length of the afore-choosen sample.

Thus to access the 5th samples 2nd electrode we simple use:
wanted_data = data_array[4]['e2']
"""

files = os.listdir(greater_dir )
data_array = []
for f in files:
    datum = load_mat_file(greater_dir + f)
    cleaned_datum = {'id': f,
                     'sng': np.array(datum['sng'][:,0], dtype = 'float64') ,
                     'e1': np.array(datum['e1'][0,:], dtype = 'float64'),
                     'e2': np.array(datum['e2'][0,:], dtype = 'float64'),
                     'ref': np.array(datum['ref'][0,:], dtype = 'float64')}
    data_array.append(cleaned_datum)
```

Logic for the various steps:

**Why do we Bandpass?**
    We bandpass to achieve two goals; firstly we remove undesired frequencies outside of our pre-determined pass ranges. Secondly, we return the signal to be centred at 0 - removing any offsets. 
    
**Why do we choose a high pass of 300 Hz?**
    This attenuates frequenies below 300 Hz in our signal. Undesirable signal properties (like a gradual shift from baseline) occur at very low frequencies when we think in terms of frequency space. Thus by removing low frequencies we remove these effects. We do wan't to be careful however, that this value isn't too high to attenuate desired characteristics of our signal.
    
**Why do we choose a low pass of 6000 Hz?**
    Here we are stating that any frequency component above this threshold is attributable to an undesired signal characteristic. In this case, the major contributing factor at >6000 Hz to the signal is noisse.
    
**Why do we Rectify?**
    Because we are measuring from a Bipolar electrode, the signal we recieve as negative voltage could be the positive signal from a neuron depending on its location from electrode tip. Thus we take the absolute of the signal - rectifying it. This removes a dimensionality of the data and forces both poles of recordings into the same dimension.
    
**Why do we Savitzky-Golay Filter?**
    Finally, we Savitzky-Golay filter to simply smooth the resulting data in a way that preserve signal to noise. A smoother trace is easily visualised, interpreted , and captures a closer estimate to what the signal itself is doing at that point in time.
    
    Because the Savitzky-Golay filter is primarily to reduce signal to noise, and smooth the data itself, it may be possible to apply a similar Butterworth Lowpass filter to fulfill the same role. This will be investigated below
    
    


```python
for f in range(len(data_array)):
    
    raw_data = abs(data_array[f]['sng']) #We rectify the raw data immediately, for visual comparison to the filtrates.
    e1_data = data_array[f]['e1']
    e2_data = data_array[f]['e2']
    
    # BANDPASS FROM 300 to 6000 HZ
    e1_data_BP = apply_BW_FILTER(e1_data, 300, 6000, 'bandpass', 1)
    e2_data_BP = apply_BW_FILTER(e2_data, 300, 6000, 'bandpass', 1)
    
    # RECTIFY
    e1_data_BP_R = abs(e1_data_BP)
    e2_data_BP_R = abs(e2_data_BP)
    
    # SAVITZKY GOLAY FILTER 50ms 4th Order Polynomial.
    window_size = 0.05 * Fs
    e1_data_BP_R_SG = apply_SG_FILTER(e1_data_BP_R, 4, window_size)
    e2_data_BP_R_SG = apply_SG_FILTER(e2_data_BP_R, 4, window_size)
    
    """ VISUALISE """
    t = np.linspace(0, len(e1_data)/Fs, len(e1_data))
    plt.figure(1, figsize = (3 * (len(e1_data) / Fs ),10))
    plt.subplot(211)
    plt.plot(t, abs(e1_data), lw = 0.5)
    plt.plot(t, e1_data_BP_R_SG, c = 'r', lw = 2)
    plt.ylim(0,7000)
    plt.xlim(1,2)
    plt.subplot(212)
    plt.plot(t, abs(e2_data), lw = 0.5)
    plt.plot(t, e2_data_BP_R_SG, c = 'r', lw = 2)
    plt.ylim(0,7000)
    plt.xlim(1,2)
    plt.show()
```
